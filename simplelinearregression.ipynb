```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("salary_data.csv")
```

```python
df
```

```python
x = pd.DataFrame(df[["Salary"]])
y = df[["YearsExperience"]]
x,y
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,
                                                 random_state = 1)
```

```python
x_train,y_train,x_test,y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
x_pred = model.predict(x_train)
y_pred = model.predict(x_test)
```

```python
x_pred
```

```python
y_pred
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(x_test,y_test)
```

```python
plt.scatter(x_train,y_train,color = "Green")
plt.plot(x_train,x_pred,color = "Red")
plt.title("Salary Vs Experience")
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.show()
```

```python
import pickle

# Save the model
with open('simplelinearregression.pkl', 'wb') as file:
    pickle.dump(model, file)




```

```python
import pickle

# Load the model
with open('simplelinearregression.pkl', 'rb') as file:
    model = pickle.load(file)
```

```python
# Example new data for prediction
new_data = pd.DataFrame({
    'Salary': [10000,20000,150000]

})

# Predict using the loaded model
experience_prediction = model.predict(new_data)

print(experience_prediction)
```
